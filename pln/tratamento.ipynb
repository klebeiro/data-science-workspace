###Carregar Dataset

###Limpeza

```python
import re
import unicodedata

def limpar_texto(texto):
    texto = texto.lower()
    texto = re.sub(r"http\S+|www\S+", "", texto)
    texto = re.sub(r"@\w+|\#", "", texto)
    texto = unicodedata.normalize("NFKD", texto).encode("ascii", "ignore").decode("utf-8")
    texto = re.sub(r"[^a-z\s]", "", texto)
    texto = re.sub(r"\s+", " ", texto).strip()
    return texto
```

```python
#aplicando no dataset
df["texto_limpo"] = df["comentario"].astype(str).apply(limpar_texto)
df.head()
```

removendo comentÃ¡rios muito curtos

```python
df = df[df["texto_limpo"].str.split().str.len() >= 3]
df.head()
```

Detectar idioma (portuguÃªs)

```python
!pip install langdetect
from langdetect import detect
```

```python
def detectar_idioma(comentario):
    try:
        return detect(comentario)
    except:
        return "unknown"

df["idioma"] = df["texto_limpo"].apply(detectar_idioma)
df = df[df["idioma"] == "pt"]

#Detectar canais BR e NÃ£o informado
df = df[
    (df["canal_pais"] == "BR") |
    (df["canal_pais"] == "Nao informado")
]
```

####Classificador 1 -  Associa_Brasil (regra)

```python
termos_brasil = [
    "brasil", "lula", "pt", "stf",
    "bolsonaro", "faz o l", "nosso pais"
]

def associa_brasil(comentario):
    return any(t in comentario for t in termos_brasil)

df["associa_brasil"] = df["texto_limpo"].apply(associa_brasil)
```

####prÃ©-filtro: prisÃ£o do Maduro

```python
termos_prisao = [
    "prender", "prisao", "buscar", "cadeia",
    "julgar", "tribunal", "punir"
]

def menciona_prisao(comentario):
    return any(t in comentario for t in termos_prisao)

df["menciona_prisao"] = df["texto_limpo"].apply(menciona_prisao)
```

####PrÃ©-filtro: intervenÃ§Ã£o EUA

```python
termos_intervencao = [
    "eua", "estados unidos", "intervencao",
    "invadir", "guerra"
]

def menciona_intervencao(comentario):
    return any(t in comentario for t in termos_intervencao)

df["menciona_intervencao"] = df["texto_limpo"].apply(menciona_intervencao)
```
