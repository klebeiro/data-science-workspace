####Buscar dados


```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/UFC /CD/pln/comentarios_tratados2.csv")

df["posicao_prisao"] = np.nan
df["posicao_intervencao"] = np.nan
```

posicao_prisao:
- APOIA_PRISAO
- CRITICA_PRISAO

posicao_intervencao:
- APOIA_INTERVENCAO_EUA
- CRITICA_INTERVENCAO_EUA


####Analisar uma amostra aleatória de 300 comentários que mencionam prisão

```python
df_prisao_anot = (
    df[df["menciona_prisao"] == True]
    .sample(300, random_state=42)
    .copy()
)

df_prisao_anot[[
    "comentario_id",
    "texto_limpo",
    "posicao_prisao"
]].head(10)
```

####Filtra comentários que mencionam intervenção

```python
df_interv_anot = (
    df[df["menciona_intervencao"] == True]
    .sample(300, random_state=42)
    .copy()
)

df_interv_anot[[
    "comentario_id",
    "texto_limpo",
    "posicao_intervencao"
]]

df_interv_sel = df_interv_anot[[
    "comentario_id",
    "texto_limpo",
    "posicao_intervencao"
]]
```

####salva os dois DataFrames em arquivos CSV no Drive

```python
df_prisao_anot.to_csv(
    "/content/drive/MyDrive/UFC /CD/pln/data/anotacao_prisao.csv", index=False
)

df_interv_anot.to_csv(
    "/content/drive/MyDrive/UFC /CD/pln/data/anotacao_intervencao.csv", index=False
)
```
