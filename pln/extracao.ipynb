```python
def video_tem_comentarios(video_id):
    try:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=1
        )
        response = request.execute()
        return True
    except:
        return False
```

```python
def buscar_videos(queries, total_videos=50):
    videos = {}

    for query in queries:
        print(f"Buscando v√≠deos para: {query}")
        next_page_token = None

        while len(videos) < total_videos:
            request = youtube.search().list(
                q=query,
                part="id,snippet",
                type="video",
                maxResults=10,
                order="viewCount",
                publishedAfter="2026-01-01T00:00:00Z",
                publishedBefore="2026-01-11T23:59:59Z",
                pageToken=next_page_token
            )

            response = request.execute()

            for item in response["items"]:
                video_id = item["id"]["videoId"]
                title = item["snippet"]["title"]

                # üîé Verifica se o v√≠deo tem coment√°rios
                if video_tem_comentarios(video_id):
                    videos[video_id] = {
                        "video_id": video_id,
                        "video_titulo": title
                    }
                    print(f"‚úîÔ∏è V√≠deo v√°lido: {title}")
                else:
                    print(f"‚ùå Coment√°rios desabilitados: {title}")

                if len(videos) >= total_videos:
                    break

            next_page_token = response.get("nextPageToken")
            if not next_page_token:
                break

        if len(videos) >= total_videos:
            break

    return list(videos.values())
```

```python
videos = buscar_videos(queries, total_videos=60)
print(f"Total de v√≠deos coletados: {len(videos)}")
print(videos[:3])  # s√≥ para conferir
```

##Extrair coment√°rios


```python
def extrair_comentarios(video_id, max_comentarios=300):
    comentarios = []
    next_page_token = None

    while len(comentarios) < max_comentarios:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=100,
            pageToken=next_page_token,
            textFormat="plainText"
        )

        response = request.execute()

        for item in response["items"]:
            snippet = item["snippet"]["topLevelComment"]["snippet"]

            comentarios.append({
                "comentario_id": item["snippet"]["topLevelComment"]["id"],
                "video_id": video_id,
                "comentario": snippet["textDisplay"],
                "comentario_data": snippet["publishedAt"]
            })

            if len(comentarios) >= max_comentarios:
                break

        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            break

        time.sleep(0.2)

    return comentarios
```

#Verifica√ß√£o

####BUSCAR DADOS DO V√çDEO (CANAL)

```python
#BUSCAR DADOS DO V√çDEO (CANAL)
def buscar_metadados_videos(video_ids):
    dados_videos = []

    for i in range(0, len(video_ids), 50):  # API aceita at√© 50 por vez
        batch = video_ids[i:i+50]

        request = youtube.videos().list(
            part="snippet",
            id=",".join(batch)
        )
        response = request.execute()

        for item in response["items"]:
            dados_videos.append({
                "video_id": item["id"],
                "canal_id": item["snippet"]["channelId"],
                "canal_titulo": item["snippet"]["channelTitle"]
            })

    return pd.DataFrame(dados_videos)
```

####BUSCAR PA√çS DO CANAL

```python
#BUSCAR PA√çS DO CANAL

def buscar_pais_canais(channel_ids):
    dados_canais = []

    for i in range(0, len(channel_ids), 50):
        batch = channel_ids[i:i+50]

        request = youtube.channels().list(
            part="snippet",
            id=",".join(batch)
        )
        response = request.execute()

        for item in response["items"]:
            dados_canais.append({
                "canal_id": item["id"],
                "canal_pais": item["snippet"].get("country", "Nao informado")
            })

    return pd.DataFrame(dados_canais)

```

```python
# Metadados dos v√≠deos
df_videos = buscar_metadados_videos(video_ids)

# Metadados dos canais
channel_ids = df_videos["canal_id"].unique().tolist()
df_canais = buscar_pais_canais(channel_ids)

# Juntar v√≠deo + canal
df_metadados = df_videos.merge(df_canais, on="canal_id", how="left")

df_metadados.head()
```

GERAR NOVA TABELA FINAL (completa)

Agora juntamos coment√°rios + metadados.

```python
df_enriquecido = df.merge(df_metadados, on="video_id", how="left")

df_enriquecido.head()
```
