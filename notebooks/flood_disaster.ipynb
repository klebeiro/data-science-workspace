### Flood disaster dataset

Extração dos dados do dataset (código dado pelo próprio site):

```python
import mlcroissant as mlc
import pandas as pd

# Fetch the Croissant JSON-LD
croissant_dataset = mlc.Dataset('https://www.kaggle.com/datasets/emanfatima2025/flood-disaster-data/croissant/download')

# Check what record sets are in the dataset
record_sets = croissant_dataset.metadata.record_sets
print(record_sets)

# Fetch the records and put them in a DataFrame
record_set_df = pd.DataFrame(croissant_dataset.records(record_set=record_sets[0].uuid))

#Faço uma limpesa nos nomes dos atributos que estavam da seguinte maneira "urban_pluvial_flood_risk_dataset.csv/city_name"
record_set_df.columns = record_set_df.columns.str.replace(r'^.*\/', '', regex=True)


record_set_df.head()
```

```python
# Extraindo a quantidade de regitros e atributos que foi pedido no documentos
record_set_df.shape
```

```python
#Retiro duas colunas das desessete para fazer as vericações
record_set_df = record_set_df.drop(columns=["dem_source", "rainfall_source"])

# Extração dos nomes das colunas (Atribultos)
record_set_df.columns
```
